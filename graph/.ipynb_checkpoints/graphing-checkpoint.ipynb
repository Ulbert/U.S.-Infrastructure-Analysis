```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
original_data = pd.read_csv("analyze.csv")
```

```python
original_data.columns
```

```python
original_data = original_data.fillna(0)
original_data = original_data.replace('\\N',0)
```

```python
def reg_line(d,x,y):#simple regression line for graph
    
    plt.plot(d[x],d[y], 'o')
    m,b = np.polyfit(d[x],d[y],1)
    plt.plot(d[x], m*d[x] + b) 
    plt.xlabel(x)
    plt.ylabel(y)
```

```python
normalized_data = pd.read_csv('analyze_normalized.csv')
normalized_data = normalized_data.fillna(0)
normalized_data = normalized_data.replace('\\N',0)
```

```python
x = normalized_data.set_index(["state", "county"])
x= x.apply(pd.to_numeric)
x= x[(x['state_local_income_tax_normalized'] < 0.4) & (x['state_local_income_tax_normalized'] > 0)]

x= x[(x['real_estate_tax_normalized'] < 0.4) & (x['real_estate_tax_normalized'] > 0)]

t = x.sample(n=100)

# t
```

```python
reg_line(t, 'state_local_income_tax_normalized', 'bridges_normalized')
```

```python
# reg_line(t, 'state_local_income_tax_normalized', 'ratio_fair_to_poor_normalized')
og = original_data.set_index(["state", "county"])
og= og.apply(pd.to_numeric)
og= og[(og['state_local_income_tax'] < 50000) & (og['state_local_income_tax'] > 0)]
# 
# x= x[(x['real_estate_tax_normalized'] < 0.4) & (x['real_estate_tax_normalized'] > 0)]

ogsamp = og.sample(n=100)
reg_line(ogsamp,'state_local_income_tax','ratio_fair_to_poor')
```
